## `sub()` funkcija

`re.sub(pattern, repl, string, count=0, flags=0)` zamjenjuje podudaranja regularnog izraza `pattern` u zadanom `string` s tekstom `repl` (ili pozivom na funkciju koja vraća zamjenski tekst).

Glavni argumenti:
- `pattern`: regularni izraz koji tražimo.
- `repl`: zamjenski niz ili funkcija koja prima `Match` objekt i vraća niz.
- `string`: input tekst u kojem se traži i zamjenjuje.
- `count`: maksimalan broj zamjena (0 znači sve).
- `flags`: opcionalni re.* flagovi (npr. `re.IGNORECASE`).

Primjer:
```python
import re
text = 'Ana ima jabuku i Ivan ima jabuke'
new = re.sub(r'jabuke?', 'kruške', text)
print(new)  # Ana ima kruške i Ivan ima kruške
```

Savjet: ako trebate složeniju logiku zamjene (npr. koristeći hvatanje grupa), proslijedite funkciju kao `repl` i u njoj obradite `Match` objekt.

````python
verse = "Marko1 Matej2 Luka3 Ivan"
re.sub('[0-9]+ ',' & ',verse)
````

````python
def normal(matchObj):
    print(matchObj.groups()) # ispis podudaranja
    if matchObj.group(1)=='-': return 'A'
    else: return 'B'

tekst = '-1234 A193 B123 B124'
re.sub('([-|A-Z])',normal,tekst)
````

## Grupe
Grupe nam služe za referenciranje uzoraka.

````python
tekst = "Srpanj 24, Kolovoz 9, Prosinac 12, Travanj D"
regex = r"[a-zA-Z]+ \d+"
re.findall(regex, tekst)

regex = r"([a-zA-Z]+) (\d+)" #ako maknemo zagradu na d npr maknut ce nam broj iz ispisa
re.findall(regex, tekst)

mObj = re.search(regex, tekst)  #pronalazi prvo podudaranje i onda staje, a findall gleda sve
mObj.group()
mObj.group(1) #samo mjesec jer pripada grupi 1
mObj.group(2) #samo dan jer pripada grupi 2
````

## Povratnost (reference grupa)
Grupe se mogu **referencirati!**

````python
# u mark down ako stavimo jednu * dovijemo ukoseno, a ako su 2 je boldano
# primjer 1: povratnost
test = 'Python&Pails'
pattern = r'([Pp])ython&\1ails'
re.match(pattern,test).group()
````

````python
# primjer 1: povratnost
test = 'Python&Pails'
regex = r'([Pp])ython&\1ails' # \1 se odnosi na prvu grupu, tj ([Pp])
re.match(regex,test).group()
````

````python
# primjer 2: 
tekst = 'zamislite bolju *Hrvatsku*, *čarobnu* Hrvatsku'
regex = r'\*(.*?)\*' # nepohlepni pristup pronalazenja *izraz* (pohlepno znaci da pokusa naci najduzi uzorak koji odgovara izrazu)... stavili regex ili pattren isto je
# \g<1> predstavlja referencu na 1. grupu: (.*?)
re.sub(regex,r'<b>\g<1></i>',tekst,count=1) # count predstavlja broj ostvarenih zamjena, samo na prvoj pojavi, \g označava grupu, a ovo 1 kao prvu
````

````python
#primjer 3
tekst = 'zamislite bolju **Hrvatsku**, **čarobnu** Hrvatsku'
boldregex = r'\*{2}(.*?)\*{2}'
italicregex = r'\*(.*?)\*' 
l1 = re.sub(boldregex,r'<b>\g<1></i>',tekst)
re.sub(italicregex,r'<i>\g<1></i>',l1)
````

## Validne e-mail adrese


````python
emails = """ime.prezime@abcd.edu;
ispravna@adresa@mail.primjer.com;
pitagora.mathos.hr"""


# rjesenje 

regex = '^\w+([\.-]?\w+)*@\w+([\.-]?\w+)*(\.\w{2,3})+$'

for e in emails.splitlines(): # razdvajanje po redovima
    email = e.rstrip(';') #maknemo s desne strane ;
    print(email,':',end=' ')
    if re.search(regex,email): 
        print("OK")  
    else:  
        print("Not OK")  
````

Implementirajte POS označavatelj koristeći n-gramski Bayesov model na primjeru Brown-ovog korpusa žanra `news`.

````python
# Setup za POS označavatelj
import nltk
import random
from nltk.corpus import brown
from nltk import RegexpTagger

# Setup Brown corpus data
brown_sents = brown.sents(categories='news')  # Get news sentences
brown_tagged_sents = [sentence for sentence in nltk.corpus.brown.tagged_sents(categories='news', tagset='universal')]

# Split data 9:1 for training and testing
size = int(0.9 * len(brown_tagged_sents))
random.shuffle(brown_tagged_sents) # permutiraj podatke 
train_sents = brown_tagged_sents[:size]
test_sents = brown_tagged_sents[size:]

# Define regexp tagger patterns
patterns = [
    (r'.*ing$', 'VERB'),  # gerunds
    (r'.*ed$', 'VERB'),   # simple past
    (r'.*es$', 'VERB'),   # 3rd singular present
    (r'.*ould$', 'VERB'), # modals
    (r'.*\'s$', 'NOUN'),  # possessive nouns
    (r'.*s$', 'NOUN'),    # plural nouns
    (r'^-?[0-9]+(.[0-9]+)?$', 'NUM'),  # cardinal numbers
    (r'.*', 'NOUN')       # nouns (default)
]

regexp_tagger = RegexpTagger(patterns)

# Test sentence for demonstration
test_sent = brown_sents[5]
tagged_sent = regexp_tagger.tag(test_sent)
````

````python
t0 = regexp_tagger # osnovni regexp parser
t1 = nltk.UnigramTagger(train_sents, backoff=t0)
t2 = nltk.BigramTagger(train_sents, backoff=t1)
t3 = nltk.TrigramTagger(train_sents,backoff=t2)


# ispisi primjere
for tok, tag in tagged_sent:
    print(tok,tag)
    
print(f'Preciznost: {t3.accuracy(test_sents)}') # udio podudarajućih vlastitih oznaka sa standardnim 
````

````python
# matrica zbunjenosti
test_tags = [tag for sent in brown.sents(categories='news')[:10] for (word, tag) in t3.tag(sent)]
gold_tags = [tag for sent in brown.tagged_sents(categories='news',tagset='universal')[:10] for (word, tag) in sent]

cm = nltk.ConfusionMatrix(gold_tags, test_tags)
print(cm)
````
