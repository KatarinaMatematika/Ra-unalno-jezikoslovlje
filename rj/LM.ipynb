## Raščlambe (tokenizacija)

```python
from nltk.tokenize import sent_tokenize, word_tokenize


sentenceEN = "Hello World. It's good to see you. Thanks for buying this book."
sentenceSP = "Hola amigo. Estoy bien."

print(sent_tokenize(sentenceEN));
print(sent_tokenize(sentenceSP));
print(word_tokenize(sentenceEN));
print(word_tokenize(sentenceSP));
```

```python
#neki izazovi kod tokenizcije
sentence = "can't its a contraction" 

from nltk.tokenize import TreebankWordTokenizer
tokenizer = TreebankWordTokenizer()
print(tokenizer.tokenize(sentence))

from nltk.tokenize import WordPunctTokenizer
tokenizer = WordPunctTokenizer()
print(tokenizer.tokenize(sentence))
```

Regularni izrazi (regex) = poseban niz znakova dani kao uzorak (engl. pattern) koji pomaže pretražiti ili pronaći podudaranje unutar stringa.
Omogućuju brzo i precizno traženje, validaciju, izdvajanje i zamjenu dijelova teksta bez ručnog parsiranja.
Pretraživanje: pronalazak uzoraka (npr. riječi, brojeva, e-mail adresa).
Validacija: provjera formata (npr. poštanski broj, telefonski broj).
Izdvajanje: izvlačenje specifičnih dijelova teksta, Primjer: izdvojite samo imena iz liste "Prezime, Ime"
Zamjena: transformacije teksta (npr. Markdown “tekst” → HTML “<b>tekst</b>”).


Automati konačnog stanja povezani su s regularnim jezicima tako da svaki takav automat prepoznaje upravo regularne jezike, odnosno skup svih nizova koje može prihvatiti prolaskom kroz svoja stanja. Oni su ekvivalentni regularnim izrazima zato što se svaki regularni izraz može pretvoriti u automat koji prihvaća isti jezik, a svaki automat može se prevesti natrag u odgovarajući regularni izraz

	Unija  L₁ ∪ L₂
Kombinacija dva jezika - stringovi koji pripadaju bilo kojem jeziku.
Primjer:
L₁ = {a, aa}
L₂ = {b, bb}
L₁ ∪ L₂ = {a, aa, b, bb}

	Konkatenacija  L_1∙L_2
Stringovi se spajaju - prvo stringovi iz L₁, zatim iz L₂.
Primjer:
L₁ = {a, aa}
L₂ = {b, bb}
L₁ · L₂ = {ab, abb, aab, aabb}

	Kleenov operator - Nula ili više ponavljanja stringa iz jezika.
Nula ponavljanja → prazan string ε
Jedno ponavljanje → string iz L
Više ponavljanja → kombinacije stringova


Regularni izrazi opisuju skup nizova koje prihvaća neki automat, dok regularne relacije opisuju preslikavanje između nizova — dakle, one ne određuju samo koji su nizovi dopušteni, nego i kako se jedan niz transformira u drugi. Drugim riječima, regularni izraz definira jezik, a regularna relacija definira odnos (ulaz → izlaz), što se najčešće ostvaruje pomoću transduktora konačnog stanja.

Transduktor konačnog stanja (FST) je proširenje konačnog automata koje umjesto “prepoznavanja” jezika vrši preslikavanje ulaznih nizova u izlazne nizove (relacije nad stringovima).
FST (transduktor konačnog stanja) sastoji se od skupa stanja, prijelaza između tih stanja, te posebno označenog početnog stanja i skupa prihvatljivih (završnih) stanja, te ulazno-izlaznih simbola na prijelazima. Služi za pretvaranje jednog niza simbola u drugi, tako da na ulazu prima niz (npr. riječ) i proizvodi odgovarajući izlaz (npr. morfološku analizu ili površinski oblik riječi).


Dvorazinska morfologija je model koji opisuje riječi kroz dvije razine — leksičku (morfeme i gramatičke značajke) i površinsku (stvarni oblik riječi u tekstu) — te pravila koja povezuju te dvije razine. U morfološkom parsiranju ona se primjenjuje tako da iz površinskog oblika riječi, primjenom tih pravila, rekonstruira njezinu leksičku strukturu i gramatička svojstva (npr. korijen, sufikse, broj, padež, vrijeme).

Jezični model je sustav koji uči obrasce iz velikih količina teksta i dodjeljuje vjerojatnosti riječima ili nizovima riječi, s ciljem da predvidi sljedeću riječ u rečenici ili procijeni koliko je neki niz riječi u jeziku vjerojatan. Uči iz korpusa (mnoštvo tekstova) obrasce pojavljivanja riječi i njihovih kombinacija. Na temelju naučenih frekvencija procjenjuje vjerojatnost nastavka (npr. koja riječ najvjerojatnije slijedi).

N-gram model procjenjuje vjerojatnost sljedeće riječi na temelju posljednjih N–1 riječi.
Primjer: u trigram modelu (N = 3) vjerojatnost sljedeće riječi ovisi o prethodne dvije.
Model računa:
P(w_n∣w_(n-1),w_(n-2),…,w_(n-N+1))

Što je N veći, model bolje “pamti” kontekst — ali zahtijeva više podataka.
N-gram modeli se temelje na Markovljevoj pretpostavci, koja kaže:
Vjerojatnost sljedeće riječi ovisi samo o ograničenom broju prethodnih riječi (lokalni kontekst), a ne o cijeloj rečenici. Model ignorira davniji kontekst jer bi ga bilo nemoguće pouzdano procijeniti iz podataka.
Zaglađivanje (smoothing) dodaje vjerojatnost neviđenim nizovima kako bi se izbjegla nula vjerojatnost, "posudi" malo vjerojatnosti od čestih kombinacija i daj je rijetkim ili neviđenim, tako da nijedna kombinacija nema potpunu nulu. Time model ostaje upotrebljiv i na novom tekstu.


Mjera za evaluaciju jezičnih modela koja pokazuje koliko je model "zbunjen" pri predviđanju riječi u jeziku.  Niža perpleksnost = bolji model. Definira se kao: PP = 2^(-log₂P(w₁...wₙ)/N) ili ekvivalentno geometrijska sredina inverznih vjerojatnosti. Koristimo ju za usporedbu različitih modela na istom test skupu i odabiremo onog s najnižom vrijednosti perpleksnosti.

Hidden Markov Model je statistički model koji se koristi za određivanje najvjerojatnijeg niza oznaka vrsta riječi (imenica, glagol, pridjev…) za zadani niz riječi u rečenici. 
Model kombinira dvije vjerojatnosti:
•	vjerojatnost prijelaza- koliko je vjerojatno da nakon jedne oznake slijedi druga
•	vjerojatnost emisije- koliko je vjerojatno da određena oznaka generira tu riječ (npr. NOUN emitira "pas")
Viterbi algoritam je dinamičko programiranje koje efikasno nalazi najvjerojatniji niz POS oznaka za danu rečenicu. Umjesto isprobavanja svih kombinacija, Viterbi pamti najbolja rješenja po koracima.
