## Raščlambe (tokenizacija)

```python
from nltk.tokenize import sent_tokenize, word_tokenize


sentenceEN = "Hello World. It's good to see you. Thanks for buying this book."
sentenceSP = "Hola amigo. Estoy bien."

print(sent_tokenize(sentenceEN));
print(sent_tokenize(sentenceSP));
print(word_tokenize(sentenceEN));
print(word_tokenize(sentenceSP));
```

```python
#neki izazovi kod tokenizcije
sentence = "can't its a contraction" 

from nltk.tokenize import TreebankWordTokenizer
tokenizer = TreebankWordTokenizer()
print(tokenizer.tokenize(sentence))

from nltk.tokenize import WordPunctTokenizer
tokenizer = WordPunctTokenizer()
print(tokenizer.tokenize(sentence))
```
